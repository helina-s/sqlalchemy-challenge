```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt 
from scipy import stats
```

```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite") 
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables 
Base.prepare(engine, reflect=True) 
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Station = Base.classes.station
Measurement = Base.classes.measurement
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
#to find list of columns in each table
inspector = inspect(engine)

measurment_columns = inspector.get_columns('measurement')
print("Measurement")
for c in measurment_columns:
    print(c['name'], c["type"])
    
station_columns = inspector.get_columns('station')
print("\nStations")
for c in station_columns:
    print(c['name'], c["type"])
```

```python
    #Precipitation Analysis
#Design a query to retrieve the last 12 months of precipitation data.
#find the last date in the database
last_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
print(f'Last date: {last_date}')

#calculate the date 1 year ago from the last date
# last_year_date = dt.date(2017,8,23) - dt.timedelta(days=365)
# print(f'Date from 1 year ago: {last_year_date}')

last_year_date = (dt.datetime.strptime(last_date[0],'%Y-%m-%d') - dt.timedelta(days=365)).strftime('%Y-%m-%d')
print(f'Date from 1 yr ago: {last_year_date}')
```

```python
#Select only the date and prcp values
query_data = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= last_year_date).all()
for record in query_data:
    print(record)
```

```python
#Load the query results into a Pandas DataFrame and set the index to the date column.
prcp_df = pd.DataFrame (query_data)
prcp_df.set_index('date', inplace=True)
prcp_df.head()
```

```python
#Sort the DataFrame values by date.
prcp_df_sort = prcp_df.sort_index()
prcp_df = prcp_df_sort
prcp_df.head()
```

```python
#Plot the results using the DataFrame plot method
ax= prcp_df.plot(rot=90, figsize=(8,6))
plt.grid(which='major', axis='both', linestyle='-')
plt.legend(['percipitation'], loc='upper right')
ax.set_ylabel("Inches")
plt.savefig('prcp_vs_date')
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
prcp_df_stat = prcp_df.describe()
prcp_df_stat.rename(columns = {"prcp": "precipitation"})
```

```python
    #Station Analysis
#Design a query to calculate the total number of stations
station_count = session.query(func.count(Station.id)).all()
```

```python
#Design a query to find the most active stations
#List the stations and the counts in descending order.
session.query(Measurement.station, func.count(Measurement.station)).\
                        group_by(Measurement.station).\
                        order_by(func.count(Measurement.station).desc()).all()
```

```python
#Which station has the highest number of observations?
highest_obs_station = session.query(Measurement.station).group_by(Measurement.station).\
              order_by(func.count(Measurement.date).desc()).first()
highest_obs_station
```

```python
#Design a query to retrieve the last 12 months of temperature observation data (TOBS).
#Filter by the station with the highest number of observations
highest_obs_data = session.query(Measurement.tobs).filter(Measurement.station=='USC00519281').\
    filter(Measurement.date>=last_year_date).all()
highest_obs_data
```

```python
#Plot the results as a histogram with bins=12
highest_obs_data_df = pd.DataFrame(highest_obs_data)
highest_obs_data_df.plot(kind='hist', bins=12)
plt.xlabel("Temprature")
plt.savefig("frequency_vs_temp.png")
```

```python
#Bonus: Other Analysis

#Temperature Analysis I
#data from June and December
june_data = session.query(Measurement.date, Measurement.tobs).\
            filter(func.strftime("%m",Measurement.date) == '06').all()
december_data = session.query(Measurement.date, Measurement.tobs).\
                filter(func.strftime("%m",Measurement.date) == '12').all()

#convert each data to DataFrame and join
june_df = pd.DataFrame(june_data, columns=["Date", "Temperature"])
december_df = pd.DataFrame(december_data, columns=["Date", "Temperature"])

june_df["Month"] = "June"
december_df["Month"] = "December"
temp_df = pd.concat([june_df, december_df], ignore_index=True)
temp_df.head()
```

```python
#Identify the average temperature in June at all stations across all available years in the dataset
#Do the same for December temperature
avg_temp = temp_df.groupby("Month").mean()
avg_temp
```

```python
#Use the t-test to determine whether the difference in the means, if any, is statistically significant
#We will use the unpaired t-test because we are comparing the average from separate gropus by month that are unrelated
#to each other.
t_test = [june_df["Temperature"],december_df["Temperature"]]
stats.ttest_ind(june_df["Temperature"], december_df["Temperature"])

#Since the p-value is less than 0.05, we can conclude that the difference in the means is statistically significant.
```

```python
    #Temperature Analysis II

# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
        return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.

start_date = '2017-05-28'
end_date = '2017-06-02'
def calc_temps(start_date, end_date):
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
temps = calc_temps(start_date, end_date)
temps
```

```python
#Plot the min, avg, and max temperature from your previous query as a bar chart.
#Use the average temperature as the bar height.
#Use the peak-to-peak (TMAX-TMIN) value as the y error bar (YERR).

df = pd.DataFrame(temps, columns=["t_min", "t_avg", "t_max"])
ax=df["t_avg"].plot.bar(yerr=(df["t_max"]-df["t_min"]), color="coral", figsize=(5,8))
plt.title("Trip Avg Temp")
plt.ylabel("Temp(F)")
ax.axes.get_xaxis().set_visible(False)
plt.show()
plt.savefig('temp_plot.png')
```

```python
#Daily Rainfall Average
#Calculate the rainfall per weather station using the previous year's matching dates.
sel = [Station.station, Station.name, func.sum(Measurement.prcp)]
rainfall = session.query(*sel).filter(Station.station == Measurement.station).\
            group_by(Measurement.station).\
            filter(Measurement.date >= start_date).\
           filter(Measurement.date <= end_date).\
           order_by(func.sum(Measurement.prcp).desc()).all()
for rain in rainfall:
    print(rain)
```

```python
#Calculate the daily normals. Normals are the averages for the min, avg, and max temperatures.
#You are provided with a function called daily_normals that will calculate the daily normals for a specific date. This
#date string will be in the format %m-%d. Be sure to use all historic TOBS that match that date string.
def daily_normals(date):
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")

startdate = dt.datetime.strptime(start_date, '%Y-%m-%d')
enddate = dt.datetime.strptime(end_date, '%Y-%m-%d')
delta = enddate -  startdate

#Create a list of dates for your trip in the format %m-%d. Use the daily_normals function to calculate the normals for
#each date string and append the results to a list.
trip_dates = []
normals = []
while (startdate <= enddate):
    trip_dates.append(startdate)
    date_str = dt.datetime.strftime(startdate, '%m-%d')
    normals.append(list(np.ravel(daily_normals(date_str))))
    startdate = startdate + dt.timedelta(days = 1)
    
normals
```

```python
#Load the list of daily normals into a Pandas DataFrame and set the index equal to the date.
daily_normals = pd.DataFrame (normals, columns=['tmin', 'tavg', 'tmax'])
daily_normals['Date'] = trip_dates
daily_normals = daily_normals.set_index('Date')
daily_normals
```

```python
#Use Pandas to plot an area plot (stacked=False) for the daily normals.
ax = daily_normals.plot.area(rot=45, stacked=False)
ax.set_ylabel("Temprature")
plt.savefig('daily_normals.png')
```
